```python
import pandas as pd

def process_dataframe(input_file):
    """
    Process Excel file and print desired columns
    
    Parameters:
    input_file (str): Path to input Excel file (.xlsx or .xls)
    """
    try:
        # Read the Excel file
        df = pd.read_excel(r"C:\Users\Parth1\.ipython\Trading journal\Gain_Loss_XB7XN_20251001_20251010.xlsx")
        
        # Add your custom columns here
        # Example transformations (modify these based on your needs):
        df['Date'] = df['Trade Date'].str.upper()
        df['calculated_column'] = df['numeric_column'] * 2
        df['combined_column'] = df['column1'] + '_' + df['column2']
        
        # Print the processed DataFrame
        print("\nProcessed Data:")
        print("===============")
        display(df)  # Using display() for better formatting in Jupyter
        
        # Print specific columns
        print("\nSelected Columns:")
        print("================")
        columns_to_show = ['uppercase_column', 'calculated_column', 'combined_column']
        display(df[columns_to_show])
        
        # Print summary statistics
        print("\nSummary Statistics:")
        print("=================")
        display(df.describe())
        
    except Exception as e:
        print(f"Error processing file: {str(e)}")

# Example usage
input_file = "path/to/your/file.xlsx"  # Replace with your Excel file path
process_dataframe(input_file)

import pandas as pd

def process_trading_journal(input_file):
    """
    Process trading journal Excel file and display key information
    
    Parameters:
    input_file (str): Path to input Excel file (.xlsx or .xls)
    """
    try:
        # Read the Excel file
        df = pd.read_excel(input_file)
        
        # Extract actual trading data (rows 8 onwards contain trade data)
        # Skip header rows (0-7) and disclaimer rows (32 onwards)
        trading_data = df.iloc[8:32].copy()
        
        # Assign proper column names based on the structure
        column_names = [
            'Script_Name', 'Security_Type', 'ISIN', 'Qty', 'Buy_Amount', 'Sell_Amount',
            'Intraday', 'Less_Than_1yr', 'Greater_Than_1yr', 'Total_Gross_PL', 'GST',
            'Brokerage', 'Misc', 'STT_CTT', 'Total_Charges', 'Net_PL', 'Gross_PL'
        ]
        trading_data.columns = column_names
        
        # Clean the data - remove rows that are totals or empty
        trading_data = trading_data.dropna(subset=['Script_Name'])
        trading_data = trading_data[~trading_data['Script_Name'].str.contains('Total', na=False)]
        
        # Convert numeric columns
        numeric_columns = ['Qty', 'Buy_Amount', 'Sell_Amount', 'Total_Gross_PL', 'GST', 
                          'Brokerage', 'Misc', 'STT_CTT', 'Total_Charges', 'Net_PL', 'Gross_PL']
        
        for col in numeric_columns:
            if col in trading_data.columns:
                trading_data[col] = pd.to_numeric(trading_data[col], errors='coerce')
        
        # Add calculated columns here (MODIFY THIS SECTION)
        # Example: Calculate profit percentage
        trading_data['Profit_Percentage'] = (trading_data['Net_PL'] / trading_data['Buy_Amount'] * 100).round(2)
        
        # Example: Categorize trades by profit/loss
        trading_data['Trade_Type'] = trading_data['Net_PL'].apply(
            lambda x: 'Profit' if x > 0 else 'Loss' if x < 0 else 'Break-even'
        )
        
        # Example: Calculate total trade value
        trading_data['Total_Trade_Value'] = trading_data['Buy_Amount'] + trading_data['Sell_Amount']

        # Calculate entry and exit prices
        trading_data['Entry_Price'] = (trading_data['Buy_Amount'] / trading_data['Qty']).round(2)
        trading_data['Exit_Price'] = (trading_data['Sell_Amount'] / trading_data['Qty']).round(2)
        
        # Display the processed data
        print("Trading Journal Analysis")
        print("=" * 50)
                
        # Show specific columns (MODIFY THIS SECTION TO CHOOSE WHAT TO DISPLAY)
        print("\nKey Metrics:") 
        print("-" * 20)
        key_columns = ['Script_Name','Buy_Amount', 'Entry_Price', 'Exit_Price', 'Profit_Percentage', 'Net_PL', 'Trade_Type']
        display(trading_data[key_columns])
        
        return trading_data
        
    except Exception as e:
        print(f"Error processing file: {str(e)}")
        return None

# Process the trading journal
trading_data = process_trading_journal(r"C:\Users\Parth1\.ipython\Trading journal\Gain_Loss_XB7XN_20251001_20251010.xlsx")
```

```python
def process_trading_data_combined(df):
    """
    Process trading data by combining related options trades (same underlying + expiry)
    This function groups options strategies together for better P&L analysis
    """
    import re
    
    def extract_underlying_and_expiry(scrip_name):
        """
        Extract underlying stock and expiry from options contract names
        Examples:
        - 'PNBHOUSING 28OCT25 XX 0' -> ('PNBHOUSING', '28OCT25')
        - 'PNBHOUSING 28OCT25 CE 920' -> ('PNBHOUSING', '28OCT25')
        - 'BHARAT ELECTRONICS LTD' -> ('BHARAT ELECTRONICS LTD', None)
        """
        # Pattern to match options: STOCK DDMMMYY OPTION_TYPE STRIKE
        options_pattern = r'^([A-Z\s]+)\s+(\d{2}[A-Z]{3}\d{2})\s+[A-Z]+\s+\d+'
        match = re.match(options_pattern, scrip_name)
        
        if match:
            underlying = match.group(1).strip()
            expiry = match.group(2)
            return underlying, expiry
        else:
            # For regular stocks, return the full name
            return scrip_name, None
    
    # Add grouping columns
    df['Underlying'] = df['Script_Name'].apply(lambda x: extract_underlying_and_expiry(x)[0])
    df['Expiry'] = df['Script_Name'].apply(lambda x: extract_underlying_and_expiry(x)[1])
    
    # Create a grouping key that combines underlying + expiry for options, or just underlying for stocks
    df['Group_Key'] = df.apply(
        lambda row: f"{row['Underlying']}_{row['Expiry']}" if row['Expiry'] else row['Underlying'], 
        axis=1
    )
    
    # Group by the new grouping key
    grouped = df.groupby('Group_Key')
    
    results = []
    for group_key, group in grouped:
        # Calculate combined amounts across all related trades
        buy_amount = group['Buy_Amount'].sum()
        sell_amount = group['Sell_Amount'].sum()
        total_charges = group['Total_Charges'].sum()
        
        # Calculate net P&L after all charges
        net_pnl = sell_amount - buy_amount - total_charges
        
        # Get the underlying name and check if it's an options strategy
        underlying = group['Underlying'].iloc[0]
        expiry = group['Expiry'].iloc[0]
        
        # Create display name
        if expiry:
            display_name = f"{underlying} {expiry} (Options Strategy)"
            # Show individual contracts in the strategy
            contracts = group['Script_Name'].unique()
            contract_details = " | ".join(contracts)
        else:
            display_name = underlying
            contract_details = ""
        
        # Calculate entry and exit prices for the main position only
        # Main position is typically the one with the largest quantity or the one that's not a hedge
        # For options strategies, we'll use the position with the highest quantity as main
        main_position = group.loc[group['Qty'].idxmax()] if len(group) > 1 else group.iloc[0]
        
        # Calculate prices based on main position only
        main_qty = main_position['Qty']
        entry_price = (main_position['Buy_Amount'] / main_qty) if main_qty > 0 else 0
        exit_price = (main_position['Sell_Amount'] / main_qty) if main_qty > 0 and main_position['Sell_Amount'] > 0 else 0
        
        # Only include if there are buy transactions
        if buy_amount > 0:
            profit_percentage = (net_pnl / buy_amount * 100) if buy_amount != 0 else 0
            
            # Determine trade status
            status = 'Open' if sell_amount == 0 else 'Closed'
            
            # Determine trade type
            trade_type = 'Profit' if net_pnl > 0 else 'Loss' if net_pnl < 0 else 'Break-even'
            
            results.append({
                'Strategy/Contract': display_name,
                'Contract_Details': contract_details,
                'Buy_Amount': buy_amount,
                'Sell_Amount': sell_amount if sell_amount > 0 else 0,
                'Entry_Price': entry_price,
                'Exit_Price': exit_price,
                'Total_Charges': total_charges,
                'Net_PnL': net_pnl,
                'Profit_Percentage': profit_percentage,
                'Status': status,
                'Trade_Type': trade_type
            })
    
    # Create final dataframe
    pnl_metrics = pd.DataFrame(results)
    
    # Round numerical columns
    numeric_columns = ['Buy_Amount', 'Entry_Price', 'Exit_Price', 'Total_Charges', 'Net_PnL', 'Profit_Percentage']
    pnl_metrics[numeric_columns] = pnl_metrics[numeric_columns].round(2)
    
    # Display results (excluding Sell_Amount column)
    print("\nCombined Trading Positions (Options Strategies Grouped):")
    print("=" * 100)
    
    # Select columns to display (excluding Sell_Amount)
    display_columns = ['Strategy/Contract', 'Contract_Details', 'Buy_Amount', 'Entry_Price', 'Exit_Price', 
                       'Profit_Percentage', 'Net_PnL', 'Status', 'Trade_Type']
    display(pnl_metrics[display_columns])
    
    # Summary statistics
    print("\nSummary:")
    print(f"Total Investment: ₹{pnl_metrics['Buy_Amount'].sum():,.2f}")
    print(f"Total Realized: ₹{pnl_metrics['Sell_Amount'].sum():,.2f}")
    print(f"Total Charges: ₹{pnl_metrics['Total_Charges'].sum():,.2f}")
    print(f"Total Net P&L: ₹{pnl_metrics['Net_PnL'].sum():,.2f}")
    
    # Separate open and closed positions
    closed_positions = pnl_metrics[pnl_metrics['Status'] == 'Closed']
    open_positions = pnl_metrics[pnl_metrics['Status'] == 'Open']
    
    if len(closed_positions) > 0:
        print(f"Closed Positions Return: {(closed_positions['Net_PnL'].sum() / closed_positions['Buy_Amount'].sum() * 100):,.2f}%")
    
    if len(open_positions) > 0:
        print(f"Open Positions Value: ₹{open_positions['Buy_Amount'].sum():,.2f}")
    
    # Show profit/loss breakdown
    profit_trades = pnl_metrics[pnl_metrics['Trade_Type'] == 'Profit']
    loss_trades = pnl_metrics[pnl_metrics['Trade_Type'] == 'Loss']
    
    print(f"\nProfit Trades: {len(profit_trades)} | Loss Trades: {len(loss_trades)}")
    if len(profit_trades) > 0:
        print(f"Average Profit: ₹{profit_trades['Net_PnL'].mean():,.2f}")
    if len(loss_trades) > 0:
        print(f"Average Loss: ₹{loss_trades['Net_PnL'].mean():,.2f}")
    
    return pnl_metrics

# Process the data with combined positions
if 'trading_data' in locals() and trading_data is not None:
    combined_metrics = process_trading_data_combined(trading_data)
else:
    print("Please run the previous cell first to load the trading data.")
```
