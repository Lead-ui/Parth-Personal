```python
import pandas as pd

def process_dataframe(input_file):
    """
    Process Excel file and print desired columns
    
    Parameters:
    input_file (str): Path to input Excel file (.xlsx or .xls)
    """
    try:
        # Read the Excel file
        df = pd.read_excel(r"C:\Users\Parth1\.ipython\Trading journal\Gain_Loss_XB7XN_20251001_20251010.xlsx")
        
        # Add your custom columns here
        # Example transformations (modify these based on your needs):
        df['Date'] = df['Trade Date'].str.upper()
        df['calculated_column'] = df['numeric_column'] * 2
        df['combined_column'] = df['column1'] + '_' + df['column2']
        
        # Print the processed DataFrame
        print("\nProcessed Data:")
        print("===============")
        display(df)  # Using display() for better formatting in Jupyter
        
        # Print specific columns
        print("\nSelected Columns:")
        print("================")
        columns_to_show = ['uppercase_column', 'calculated_column', 'combined_column']
        display(df[columns_to_show])
        
        # Print summary statistics
        print("\nSummary Statistics:")
        print("=================")
        display(df.describe())
        
    except Exception as e:
        print(f"Error processing file: {str(e)}")

# Example usage
input_file = "path/to/your/file.xlsx"  # Replace with your Excel file path
process_dataframe(input_file)

import pandas as pd

def process_trading_journal(input_file):
    """
    Process trading journal Excel file and display key information
    
    Parameters:
    input_file (str): Path to input Excel file (.xlsx or .xls)
    """
    try:
        # Read the Excel file
        df = pd.read_excel(input_file)
        
        # Extract actual trading data (rows 8 onwards contain trade data)
        # Skip header rows (0-7) and disclaimer rows (32 onwards)
        trading_data = df.iloc[8:32].copy()
        
        # Assign proper column names based on the structure
        column_names = [
            'Script_Name', 'Security_Type', 'ISIN', 'Qty', 'Buy_Amount', 'Sell_Amount',
            'Intraday', 'Less_Than_1yr', 'Greater_Than_1yr', 'Total_Gross_PL', 'GST',
            'Brokerage', 'Misc', 'STT_CTT', 'Total_Charges', 'Net_PL', 'Gross_PL'
        ]
        trading_data.columns = column_names
        
        # Clean the data - remove rows that are totals or empty
        trading_data = trading_data.dropna(subset=['Script_Name'])
        trading_data = trading_data[~trading_data['Script_Name'].str.contains('Total', na=False)]
        
        # Convert numeric columns
        numeric_columns = ['Qty', 'Buy_Amount', 'Sell_Amount', 'Total_Gross_PL', 'GST', 
                          'Brokerage', 'Misc', 'STT_CTT', 'Total_Charges', 'Net_PL', 'Gross_PL']
        
        for col in numeric_columns:
            if col in trading_data.columns:
                trading_data[col] = pd.to_numeric(trading_data[col], errors='coerce')
        
        # Add calculated columns here (MODIFY THIS SECTION)
        # Example: Calculate profit percentage
        trading_data['Profit_Percentage'] = (trading_data['Net_PL'] / trading_data['Buy_Amount'] * 100).round(2)
        
        # Example: Categorize trades by profit/loss
        trading_data['Trade_Type'] = trading_data['Net_PL'].apply(
            lambda x: 'Profit' if x > 0 else 'Loss' if x < 0 else 'Break-even'
        )
        
        # Example: Calculate total trade value
        trading_data['Total_Trade_Value'] = trading_data['Buy_Amount'] + trading_data['Sell_Amount']

        # Calculate entry and exit prices
        trading_data['Entry_Price'] = (trading_data['Buy_Amount'] / trading_data['Qty']).round(2)
        trading_data['Exit_Price'] = (trading_data['Sell_Amount'] / trading_data['Qty']).round(2)
        
        # Display the processed data
        print("Trading Journal Analysis")
        print("=" * 50)
                
        # Show specific columns (MODIFY THIS SECTION TO CHOOSE WHAT TO DISPLAY)
        print("\nKey Metrics:") 
        print("-" * 20)
        key_columns = ['Script_Name','Buy_Amount', 'Entry_Price', 'Exit_Price', 'Net_PL', 'Profit_Percentage', 'Trade_Type']
        display(trading_data[key_columns])
        
        return trading_data
        
    except Exception as e:
        print(f"Error processing file: {str(e)}")
        return None

# Process the trading journal
trading_data = process_trading_journal(r"C:\Users\Parth1\.ipython\Trading journal\Gain_Loss_XB7XN_20251001_20251010.xlsx")
```
